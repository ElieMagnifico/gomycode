```python
############# Preprocessing part ##############
import pandas as pd
print("******Here is our preprocessing part********\n")
dataset=pd.read_csv("titanic-passengers.csv",encoding="ISO-8859-1")
print(dataset.head())
print("\n",dataset.info())
print("\n",dataset.isnull().sum)
dataset['Age'].fillna(dataset['Age'].mean(), inplace=True)
vac=dataset['Cabin'].value_counts()
print(vac)
D1=dict(vac)
j=0
a=None
for i in D1.values():
    if i>j:
        j=i
for n in D1.keys():
    if D1[n]==j:
        a=n
        dataset['Cabin'].fillna(a,inplace=True)
print("\n",dataset)
if dataset.isnull().sum().sum()==0:
    print("\n The data is ready to be treated ! ðŸ™‚")
else :
    print("\n The preprocessing is not finished yet")

############# Visualization part ##############
print("\n *******This is our visualization part********* \n\n")
import matplotlib.pyplot as plt
import seaborn as sns
plt.title("Histogram of different ages")
plt.xlabel("Age")
dataset['Age'].plot.hist()
plt.show()
plt.title("Bar plot of cabins")
plt.xlabel("Cabin")
vac.plot.bar(rot=45)
plt.show()
nbr=dataset['Sex'].value_counts()
plt.title("Bar plot of sex")
plt.xlabel("Sex")
nbr.plot.bar(rot=-45)
plt.show()
grid=sns.FacetGrid(dataset,col='Survived',height=2.2,aspect=1.6)
grid.map(sns.barplot,'Age','Sex',order=['male','female'],alpha=.5,ci=None,color='green')
grid.add_legend()
grid=sns.FacetGrid(dataset,col='Survived',height=2.2,aspect=1.6)
grid.map(sns.barplot,'Pclass','Embarked',order=["C", "S","Q"],alpha=.5,ci=None,color='yellow')
grid.add_legend()
def plot_correlation_map( df ):
    corr = df.corr()
    s , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
        )
    return s
print(plot_correlation_map(dataset))
print("This function plots an heatmap according to the given arguments: -corr is the name of the data, -cmap takes the values of the data\n",
      "to color the space, -square sets the axes aspect, -cbar_kws is the argument for the figure colorbar, -ax takes the axes in which to\n",
      "draw the plot, -annot writes the data values in each cell ( value 1 in this case), -annot_kws takes the keyword argument for axes text")
```
